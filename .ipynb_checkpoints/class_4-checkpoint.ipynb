```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as scp

eq = lambda x: x**2 - 3 * x + 2
```

```python
xvalues = np.linspace(0,3, 301)
plt.plot(xvalues, eq(xvalues), label = r"$f(x) = x^2 - 3x + 2$")
plt.axhline(0,color = 'gray', linestyle = '--')
plt.legend()
```

```python
from scipy.optimize import minimize
```

```python
obj_eq = lambda x: abs(x**2 - 3*x + 2)
```

```python
for strt in [0,4]:
    result = minimize(
        fun = obj_eq,
        x0 = strt,
        method = 'Nelder-Mead',
        tol = 1e-6
    )
    
    print(result)
```

```python
def objectiv_eq(x, Kp):
    return abs((2 + x) * (2 - 2*x)**2 * Kp - 4*x**3)
```

```python
result = minimize(
    fun = objectiv_eq,
    x0 = 0,
    args = (10.060,),
    method = 'Nelder-Mead',
    tol = 1e-6
)
print(result['x'])
```
